# Text file generator

```python
image_files = []
image_folder_path="/content/drive/MyDrive/Yolov4/darknet/build/darknet/x64/data/hardhat/Test_images/Images"
text_file_path= "/content/drive/MyDrive/Yolov4/darknet/build/darknet/x64/data/hardhat/Test_images/label/Test.txt"
for filename in os.listdir(image_folder_path):
    if filename.endswith(".png"):
        complete_path=os.path.join(image_folder_path,filename)
        image_files.append(complete_path)
        #print(filename)
        
with open(text_file_path, "w") as outfile:
    for image in image_files:
        outfile.write(image)
        outfile.write("\n")
    outfile.close()
```
